```python
import etl
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = 'cora'
s1 = 'snap'
fp = '/../Users/aureliobarrios/School/fall2020/dsc180/DSC180A-Checkpoint1/data/raw/cora/cora.content'
fp2 = '/../Users/aureliobarrios/School/fall2020/dsc180/DSC180A-Checkpoint1/data/raw/cora/cora.cites'
fp3 = '/../Users/aureliobarrios/School/fall2020/dsc180/DSC180A-Checkpoint1/data/raw/twitch/ENGB/musae_ENGB_edges.csv'
fp4 = '/../Users/aureliobarrios/School/fall2020/dsc180/DSC180A-Checkpoint1/data/raw/twitch/ENGB/musae_ENGB_features.json'
```

```python
d1 = etl.read_edges(s1, fp3)
d2 = etl.read_features(s1, fp4)
```

```python
id_nodes = etl.index_nodes(d2)
```

```python
adj = etl.create_adj(id_nodes, d1)
```

```python
G = nx.from_numpy_matrix(adj, nx.DiGraph())
plt.figure(figsize=(10,10))
nx.draw(G, node_size=10, edge_size=1)
plt.show()
```
